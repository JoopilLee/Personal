# 연도 뽑기

```python
# step1. 데이터 불러와서 띄어쓰기 기준으로 단어 분리
data1_years = open('King_Yeonsan.txt', encoding = 'utf-8').read()
data2_years = []

for i in data1_years.split('연산군일기'):
    data2_years.append(i)
data2_years = data2_years[1:]
data2_years[0]
```

```python
# 연도만 뽑아주기
import re
p = re.compile('[0-9]{4}')
data2_year = []
for d in range(len(data2_years)):
    data2_year.append("".join(p.findall(data2_years[d])))
```

```python
for d in range(len(data2_year)):
    if len(data2_year[d]) > 4:
        data2_year[d] = data2_year[d][:4]
data2_year
```

# 분류 뽑기

```python
# step1. 데이터 불러와서 띄어쓰기 기준으로 단어 분리
data1_gubun = open('King_Sejong.txt', encoding = 'utf-8').read()
data2_gubun = []

for i in data1_gubun.split('세종실록'):
    data2_gubun.append(i)
data2_gubun = data2_gubun[1:]
data2_gubun[0]
```

```python
# 분류부터 뒤까지
for d in range(len(data2_gubun)):
    data2_gubun[d] = data2_gubun[d][data2_gubun[d].find('【분류】'):]
data2_gubun[3]
```

```python
# 분류 제거
for d in range(len(data2_gubun)):
    data2_gubun[d] = data2_gubun[d][data2_gubun[d].find('\n')+1:]
```

```python
import re
s = []
try:
    for d in range(len(data2_gubun)):
        for t in re.finditer('\n', data2_gubun[d]):
            s.append(t.start()) 
        data2_gubun[d] = data2_gubun[d][:s[1]-1]   
        s = []
except:
    data2_gubun[d]
data2_gubun
```

```python
# 위에서 except에 걸린 친구들
for d in range(len(data2_gubun)):
    if data2_gubun[d].find('[') != -1:
        data2_gubun[d] = data2_gubun[d][:data2_gubun[d].find('[')-1]
```

# 분류 1개만 뽑기

```python
# /기준으로 split해서 리스트로 만들어주기
for d in range(len(data2_gubun)):
    data2_gubun[d] = data2_gubun[d].split('/')
data2_gubun[:5]
```

```python
# 중분류만 뽑아주기
for d in range(len(data2_gubun)):
    for dd in range(len(data2_gubun[d])):
        data2_gubun[d][dd] = data2_gubun[d][dd].split('-')[0]
data2_gubun[:5]
```

```python
# 한자 지워주기
import re 
p = re.compile('[가-힣]{1,4}')

for d in range(len(data2_gubun)):
    for dd in range(len(data2_gubun[d])):
        data2_gubun[d][dd] = p.findall(data2_gubun[d][dd])
```

```python
# 리스트 풀어주기
for d in range(len(data2_gubun)):
    data2_gubun[d] = sum(data2_gubun[d],[])
data2_gubun[3]
```

```python
# 빈 곳 없는지 확인하기
for d in range(len(data2_gubun)):
    if data2_gubun[d] == []:
        print(d)
```

```python
# 분류만 전체다 뽑기
bun = []
for d in data2_gubun:
    for dd in d:
        bun.append(dd)
```

```python
# 분류만 따로 txt파일 만들어주기
for b in bun:
    with open("세종분류색인_중.txt",'a',encoding='utf-8') as file:
                            file.write(b)
                            file.write('\n')
```
