# 분류색인 - 대

```python
# 데이터 불러와서 단어별로 리스트에 저장
import pandas as pd
ys = open('연산군분류색인_중.txt', encoding = 'utf-8').read()
jj = open('세종분류색인_중.txt', encoding = 'utf-8').read()
ys_big, sj_big = ys.split('\n'), sj.split('\n')
```

```python
# 대분류 라벨링
politics = ['왕실','정론','행정','인사','사법','군사','외교','변란','인물']
economy = ['재정','금융','물가','상업','무역','교통','도량형','농업','수산업','광업','공업','건설']
social = ['가족','호구','신분','향촌','의생활','식생활','주생활','윤리','풍속','구휼','보건']
culture = ['사상','역사','과학','의약','어문학','예술','교육','출판']

for y in range(len(ys_big)):
    if ys_big[y] in politics:
        ys_big[y] = '정치'
    elif ys_big[y] in economy:
        ys_big[y] = '경제'
    elif ys_big[y] in social:
        ys_big[y] = '사회'
    else:
        ys_big[y] = '문화'
```

```python
# 정조도 대분류로 변환
for j in range(len(sj_big)):
    if sj_big[j] in politics:
        sj_big[j] = '정치'
    elif sj_big[j] in economy:
        sj_big[j] = '경제'
    elif sj_big[j] in social:
        sj_big[j] = '사회'
    else:
        sj_big[j] = '문화'
```

```python
ys_count = Counter(ys_big).most_common()
sj_count = Counter(sj_big).most_common()
```

```python
ind = list(dict(ys_count).keys())
ys1 = list(dict(ys_count).values())
sj1 = list(dict(sj_count).values())

df = pd.DataFrame({'연산' : ys1, '세종' : sj1}, index = ind)
df
```

```python
import warnings
import numpy as np
warnings.filterwarnings('ignore')
plt.rcParams['font.family'] = 'AppleGothic'

fig, ax = plt.subplots(figsize=(12,6))
bar_width = .35
index = np.arange(4)

plt.title('연산군-세종 대분류', size=25)
b1 = plt.bar(index, df['연산'], bar_width, alpha =.75, label = '연산군', color = 'red')
b2 = plt.bar(index + bar_width, df['세종'], bar_width, alpha =.75, label = '세종', color = 'blue')
plt.xticks(np.arange(bar_width, 4 + bar_width, 1), ind)

plt.xlabel('대분류', size = 13)
plt.ylabel('count', size = 13)
plt.legend() 
plt.show()
```

# 분류색인 - 소

```python
# step1. 데이터 불러와서 띄어쓰기 기준으로 단어 분리
data1_gubun = open('King_Sejong.txt', encoding = 'utf-8').read()
data2_gubun = []

for i in data1_gubun.split('세종실록'):
    data2_gubun.append(i)
data2_gubun = data2_gubun[1:]
data2_gubun[0]

# 분류부터 뒤까지
for d in range(len(data2_gubun)):
    data2_gubun[d] = data2_gubun[d][data2_gubun[d].find('【분류】'):]

# 분류 제거
for d in range(len(data2_gubun)):
    data2_gubun[d] = data2_gubun[d][data2_gubun[d].find('\n')+1:]

import re
s = []
try:
    for d in range(len(data2_gubun)):
        for t in re.finditer('\n', data2_gubun[d]):
            s.append(t.start()) 
        data2_gubun[d] = data2_gubun[d][:s[1]-1]   
        s = []
except:
    data2_gubun[d]

# 위에서 except에 걸린 친구들
for d in range(len(data2_gubun)):
    if data2_gubun[d].find('[') != -1:
        data2_gubun[d] = data2_gubun[d][:data2_gubun[d].find('[')-1]
# /기준으로 split해서 리스트로 만들어주기
for d in range(len(data2_gubun)):
    data2_gubun[d] = data2_gubun[d].split('/')

# 소분류만 뽑기
import re 
p = re.compile('-[가-힣]{1,4}')

for d in range(len(data2_gubun)):
    for dd in range(len(data2_gubun[d])):
        data2_gubun[d][dd] = p.findall(data2_gubun[d][dd])

# 리스트 풀어주기
for d in range(len(data2_gubun)):
    data2_gubun[d] = sum(data2_gubun[d],[])
data2_gubun[3]

# 앞에 - 지워주기
for d in range(len(data2_gubun)):
    for dd in range(len(data2_gubun[d])):
        data2_gubun[d][dd] = data2_gubun[d][dd][1:]

# 빈 곳 없는지 확인하기
# for d in range(len(data2_gubun)):
#     if data2_gubun[d] == []:
#         print(d)

# 분류만 전체다 뽑기
bun = []
for d in data2_gubun:
    for dd in d:
        bun.append(dd)

# 분류만 따로 txt파일 만들어주기
for b in bun:
    with open("세종분류색인_소.txt",'a',encoding='utf-8') as file:
                            file.write(b)
                            file.write('\n')
```
